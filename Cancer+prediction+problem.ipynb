```python
print(inputdf.loc[:,[1,2,3,20,21]].head(10))
```

```python
inputdf[[20,21]] = inputdf[[20,21]].apply(pd.to_numeric, errors='coerce', axis=0)
```

```python
inputlabelsdf = pd.get_dummies(inputdf,dummy_na=True)
```

```python
inputlabelsdf.head()
```

```python
inputlabelsdf.info()
```

```python
NUMERIC= list(range(4,34))
```

```python
print(NUMERIC)
```

```python
from sklearn.preprocessing import Imputer
```

```python
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
```

```python
inputlabelsdf[NUMERIC]= imp.fit_transform(inputlabelsdf[NUMERIC])
```

```python
inputlabelsdf.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputlabelsdf,outputdf,test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
print 'Accuracy with logistic regression: ' + str(logreg.score(X_test,y_test)*100) + ' %'
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(X_train,y_train)
```

```python
print 'Accuracy with knn: ' + str(knn.score(X_test, y_test)*100) + ' %'
```

```python
from sklearn import svm
```

```python
classifier=svm.SVC()
```

```python
classifier.fit(X_train,y_train)
```

```python
print 'Accuracy with SVM: ' + str(classifier.score(X_test, y_test)*100) + ' %'
```
